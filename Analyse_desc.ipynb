```python
import pandas as pd

# Charger le fichier trust.txt
trust_data = pd.read_csv('dl_trust.txt', sep=' ', header=None, names=['user_id', 'trust_user_id', "trust"])

# Charger le fichier rating.txt
rating_data = pd.read_csv('dl_ratings.txt', sep=' ', header=None, names=['user_id', 'item_id', 'rating'])
```

```python
n_users = trust_data['user_id'].nunique()
n_interactions = len(trust_data)

print(f"Nombre d'utilisateurs : {n_users}")
print(f"Nombre d'interactions : {n_interactions}")
```

```python
density = n_interactions / (n_users * (n_users - 1))
print(f"Densité : {density}")
```

```python
import matplotlib.pyplot as plt

# Distribution des évaluations
rating_data['rating'].hist()
plt.title("Distribution des Ratings")
plt.xlabel("Rating")  # Ajouter un label pour l'axe x
plt.ylabel("Nombre d'Utilisateurs")  # Ajouter un label pour l'axe y
plt.show()

# Distribution des relations de confiance
trust_data.groupby('user_id').size().hist()
plt.title("Distribution des Relations de Confiance par Utilisateur")
plt.xlabel("Nombre de Relations de Confiance")  # Ajouter un label pour l'axe x
plt.ylabel("Nombre d'Utilisateurs")  # Ajouter un label pour l'axe y
plt.show()
```

```python
sparsity = 1 - (rating_data.groupby(['user_id', 'item_id']).size().sum() / (n_users * rating_data['item_id'].nunique()))
print(f"Sparsité : {sparsity}")
```
